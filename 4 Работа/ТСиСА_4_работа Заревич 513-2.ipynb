# 2 Жертвы

```python
# В наборе данных имеется информация о Возрасте, Поле каждой жертвы. Есть ли связь между этими признаками? Для этого рассчитайте корреляционную матрицу.
import numpy as np

# кодирование категориальных значений для столбца Victim Sex
sex_encoded = pd.get_dummies(data["Vict Sex"], prefix="Sex")

# объединение закодированных значений с числовыми значениями
matrix = pd.concat([data["Vict Age"], sex_encoded], axis=1)

# вычисление корреляционной матрицы
corr_matrix = matrix.corr()

# вывод корреляционной матрицы
print(corr_matrix)
```

```python
# Дайте ответ на вопрос: верно ли, что женщины чаще оказываются жертвами по сравнению с мужчинами?
female_data = data.loc[data['Vict Sex'] == 'F']

male_data = data.loc[data['Vict Sex'] == 'M']

print("Мужчины-жертвы: ", male_data.shape[0])
print("Женщины-жертвы: ", female_data.shape[0])

# Вывод: мужчины оказываются жертвами чаще
```

# 3 Преступления, пол и возраст

```python
# Изучите распределение количества преступлений по возрасту. Для этого постройте гистограмму по переменной возраст.
# Какова тенденция? Люди какого возраста чаще всего подвергаются преступлениям? Есть ли локальные минимумы?
# Убираем пустые клетки
data = data[(data['Vict Age'] > 0) & (data['Vict Age'] < 120)]

# 1. Гистограмма
plt.figure(figsize=(12, 6))
data['Vict Age'].hist(bins=30, color='blue', edgecolor='white', grid=False)
plt.title('Распределение преступлений по возрасту жертв', fontsize=14)
plt.xlabel('Возраст жертвы (Vict Age)', fontsize=12)
plt.ylabel('Количество преступлений', fontsize=12)
plt.show()

# 2. KDE
plt.figure(figsize=(12, 6))
data['Vict Age'].plot.kde(color='red', linewidth=3)
plt.title('Плотность распределения по возрасту', fontsize=14)
plt.xlabel('Возраст жертвы (Vict Age)', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.grid(True, alpha=0.2)
plt.show()

# 3. Комбинированный график
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.hist(data['Vict Age'], bins=30, color='royalblue', alpha=0.7, density=True)
ax1.set_xlabel('Возраст жертвы (Vict Age)', fontsize=12)
ax1.set_ylabel('Нормализованная частота', fontsize=12)

ax2 = ax1.twinx()
data['Vict Age'].plot.kde(ax=ax2, color='darkred', linewidth=3)
ax2.set_ylabel('Плотность вероятности', fontsize=12)

plt.title('Нормализованная гистограмма и KDE', fontsize=14)
plt.show()

# Вывод статистики
print("Анализ возраста жертв:")
print(f"Средний возраст: {data['Vict Age'].mean():.1f} лет")
print(f"Медианный возраст: {data['Vict Age'].median():.1f} лет")
print(f"Наиболее частый возраст: {data['Vict Age'].mode()[0]} лет")
print(f"Стандартное отклонение: {data['Vict Age'].std():.1f} лет")
```

```python
# Фильтруем только мужчин и женщин
filtered_data = data[data['Vict Sex'].isin(['M', 'F'])]


# Разделяем данные по полу
male_data = filtered_data[(filtered_data['Vict Sex'] == 'M') & (data['Vict Age'] > 0)]
female_data = filtered_data[(filtered_data['Vict Sex'] == 'F') & (data['Vict Age'] > 0)]

# Создаем фигуру с двумя подграфиками
plt.figure(figsize=(14, 6))

# Гистограмма для мужчин
plt.subplot(1, 2, 1)
male_data['Vict Age'].hist(bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Распределение возраста жертв (Мужчины)')
plt.xlabel('Возраст')
plt.ylabel('Количество преступлений')
plt.grid(axis='y', alpha=0.3)

# Гистограмма для женщин
plt.subplot(1, 2, 2)
female_data['Vict Age'].hist(bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title('Распределение возраста жертв (Женщины)')
plt.xlabel('Возраст')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
